```python
import nltk
```

```python
import urllib
import bs4 as bs
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# getting the data source


source=urllib.request.urlopen('https://en.wikipedia.org/wiki/Global_warming')
```

```python
#parsing the data by creating Beautiful soap object
soap=bs.BeautifulSoup(source,'lxml')
```

```python
#Fetching the data
text=''
for paragraph in soap.find_all('p'):
    text+=paragraph.text
```

```python
text
```

```python
# Preprocessing the data

text=re.sub(r'\[[0-9]*\]',' ',text)
text=text.lower()
text=re.sub(r'\n','',text)
text=re.sub(r'\d',' ',text)
text=re.sub(r"\'s+",'',text)
```

```python
text
```

```python
#Preparing the data set

sentance=nltk.sent_tokenize(text)
```

```python
sentance
```

```python
# importing pandas dataframe 

import pandas as pd
```

```python
data=pd.DataFrame(sentance,columns=['Sentance'])
```
